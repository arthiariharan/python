```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['Kilometers_Driven', 'Fuel_Type', "Transmission", "Owner_Type"])
y = np.array([3, 8, 1, 10])
colors=["#4CAF50","red","blue","orange"]
plt.bar(x, y, color=colors)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Example 1: Distribution of car prices
plt.figure(figsize=(10, 6))
sns.histplot(car['Price'], bins=20, kde=True)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'Type' is the column representing car types (update if necessary)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Mileage', y='Price', data=car)
plt.title('Boxplot of Car Prices by Car Type')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
# Set background color
plt.gcf().set_facecolor('green')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'car' is your DataFrame containing the data
plt.figure(figsize=(10, 10))
# Define a palette using a list of colors
custom_palette = ["#FF0000", "#00FF00", "#0000FF"]  # Example hexadecimal colors

# Set the palette
sns.set_palette(custom_palette)

# Scatter plot with custom palette
sns.scatterplot(x='Brand', y='Power',hue="Transmission", data=car, palette=custom_palette)
plt.title('Brand wise Power')
plt.xlabel("Brand")
plt.ylabel('Power')  
plt.show()

```

```python
import matplotlib.pyplot as plt
# Line plot of car prices over time
plt.figure(figsize=(10, 6))
car.groupby('Year')['Price'].mean().plot(kind='line')
plt.title('Average Car Price Over Time')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Brand', y='Mileage', data=car)

# Add title and labels
plt.title('Average Mileage by Car Brand')
plt.xlabel('Brand')
plt.ylabel('Mileage')

# Rotate x-axis labels for better readability

plt.xticks(rotation=45)

# Show the plot
plt.tight_layout() # Adjust layout to fit everything cleanly
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot
plt.hist(data, bins=30)
plt.title('Histogram')
plt.xlabel('Brand')
plt.ylabel('Kilometers_Driven')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Model", "Engine", "Mileage", "Brand"]
myexplode = [0.2, 0, 0, 0]

# Define custom colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Example colors (blue, orange, green, red)

plt.pie(y, labels=mylabels, explode=myexplode, shadow=True, colors=colors)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(x='Brand', y='Price', data=car, palette='viridis')
plt.title('Average Price of Cars by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
sns.pairplot(car[['Year','Kilometers_Driven','Mileage', 'Engine', 'Power', 'Seats', 'Price' ]], diag_kind='kde')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 6), dpi=200)
# List of columns for correlation analysis
column_list = ['Engine', 'Year', 'Mileage', 'Price']
# Create a heatmap of the correlation matrix for the specified columns
sns.heatmap(car[column_list].corr(), annot=True)
# Display the plot
plt.show()
```
